```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

```python
a1 = pd.read_csv('a1', index_col=0)
a1
```

```python
a2 = pd.read_csv('a2', index_col=0)
a2
```

```python
a3 = pd.read_csv('a3', index_col=0)
a3
```

```python
a4 = pd.read_csv('a4', index_col=0)
a4
```

```python
all_fitness = pd.concat([a1[['alg_f']], a2['alg_f'], a3['alg_f'], a4['alg_f']], axis=1)
all_fitness.dropna(inplace=True)
all_fitness.columns = ['alg_f_1', 'alg_f_2', 'alg_f_3', 'alg_f_4']
all_fitness
```

```python
all_fitness.plot()
```

```python
all_fitness = pd.concat([a2['alg_f'], a1['gen_f'], a2['gen_f'], a3['gen_f'], a4['gen_f']], axis=1)
all_fitness.dropna(inplace=True)
all_fitness.columns = ['alg_f', 'exp_1', 'exp_2', 'exp_3', 'exp_4']
all_fitness
```

```python
ax = all_fitness.plot()

# for column in all_fitness.columns:
#     x = all_fitness.index[-1]  # X position at the end of the series
#     y = all_fitness[column].iloc[-1]  # Y position at the last point of the series
#     print(x, y)
#     ax.annotate(column, xy=(x, y), xytext=(5, 0), textcoords='offset points', ha='left', va='center')

ax.set_title('Fitness por tamanho da Ã¡rvore')
ax.set_xlabel('|V(T)|')
ax.set_ylabel('Fitness')
plt.show()

# Export the plot to a file (e.g., PNG)
ax.get_figure().savefig('all_fitness.png')
```

```python
all_time = pd.concat([a2['alg_t'], a1['gen_t'], a2['gen_t'], a3['gen_t'], a4['gen_t']], axis=1)
all_time.dropna(inplace=True)
all_time.columns = ['alg_t', 'exp_1', 'exp_2', 'exp_3', 'exp_4']
all_time
```

```python
ax = all_time.plot()

ax.set_title('Tempo por tamanho da Ã¡rvore')
ax.set_xlabel('|V(T)|')
ax.set_ylabel('Tempo (s)')
plt.show()
# Export the plot to a file (e.g., PNG)
ax.get_figure().savefig('all_time.png')
```
